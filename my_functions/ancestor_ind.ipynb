```python
def create_ancestor_individual(num_chromosomes: int, num_loci_per_chr: int, allele_type) -> 'Individual':
    """
    Creates a single homozygous ancestor 'Individual' with all loci set to a specified allele type.

    This function is fundamental for setting up the initial parental populations. It ensures
    that each new ancestor individual has all its chromosomes and all loci on those chromosomes
    fixed to either Allele0 (0), Allele1 (1), or Allele2 (2), representing a pure ancestral line.

    Args:
        num_chromosomes (int): The number of diploid chromosome pairs this ancestor will have.
        num_loci_per_chr (int): The number of loci on each chromosome.
        allele_type (Union[str, int]): The desired allele type to populate the ancestor's genome.
                                       Can be an integer (0, 1, 2) or a string ('0', '1', '2', 'A0', 'A2').

    Returns:
        Individual: A newly created 'Individual' object, fully homozygous for the specified allele type.

    Raises:
        ValueError: If an unsupported string or integer value is provided for `allele_type`.
        TypeError: If `allele_type` is not a string or an integer.
    """
    # Validate and convert the input `allele_type` to its internal integer representation (0, 1, or 2).
    # This flexible input handling allows for calling the function with different convenient formats.
    if isinstance(allele_type, str):
        if allele_type == '0' or allele_type.upper() == 'A0':
            initial_allele_int = 0
        elif allele_type == '1': # Assuming '1' might represent a fixed heterozygous state, or another pure line
            initial_allele_int = 1
        elif allele_type == '2' or allele_type.upper() == 'A2':
            initial_allele_int = 2
        else:
            raise ValueError(f"Unsupported allele_type string: {allele_type}. Expected '0', '1', '2', 'A0', or 'A2'.")
    elif isinstance(allele_type, int):
        if allele_type in (0, 1, 2):
            initial_allele_int = allele_type
        else:
            raise ValueError(f"Unsupported allele_type integer: {allele_type}. Expected 0, 1, or 2.")
    else:
        raise TypeError(f"allele_type must be str or int, got {type(allele_type)}")

    # Create a new 'Individual' instance, which will get a unique global ID.
    individual = Individual(
        num_chromosomes=num_chromosomes,
        num_loci_per_chromosome=num_loci_per_chr
    )

    # Populate the individual's chromosomes. Ancestors are typically homozygous,
    # meaning both homologous chromosomes (chromatid1 and chromatid2) for each pair
    # will have identical alleles across all loci.
    for _ in range(num_chromosomes):
        # Create a list of alleles where every locus has the `initial_allele_int`.
        chromosome_alleles = [initial_allele_int] * num_loci_per_chr

        # Create two 'Chromosome' objects, both carrying identical alleles, for the diploid pair.
        # `[:]` creates a shallow copy to ensure they are distinct objects, though their content is the same.
        chromatid1 = Chromosome(chromosome_alleles[:])
        chromatid2 = Chromosome(chromosome_alleles[:])

        # Add the newly created 'DiploidChromosomePair' to the individual's list of chromosome pairs.
        individual.diploid_chromosome_pairs.append(DiploidChromosomePair(chromatid1, chromatid2))

    return individual
```

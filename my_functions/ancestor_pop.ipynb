```python
def create_ancestor_populations(
    num_individuals: int,
    num_chromosomes: int,
    num_loci_per_chr: int,
    allele_type
) -> list['Individual']:
    """
    Creates a population of multiple homozygous ancestor individuals.

    This function acts as a wrapper around `create_ancestor_individual`,
    calling it multiple times to generate a list of pure ancestral individuals,
    all with the same specified allele type.

    Args:
        num_individuals (int): The desired number of individuals in this ancestral population.
        num_chromosomes (int): The number of diploid chromosome pairs for each individual.
        num_loci_per_chr (int): The number of loci on each chromosome for each individual.
        allele_type (Union[str, int]): The allele type (0, 1, 2, 'A0', 'A2') that all individuals
                                       in this population will be homozygous for.

    Returns:
        List[Individual]: A list containing the newly created ancestral 'Individual' objects.
    """
    # Uses a list comprehension to efficiently create `num_individuals` instances
    # of 'Individual', each fully homozygous for the `allele_type`.
    return [create_ancestor_individual(num_chromosomes, num_loci_per_chr, allele_type) for _ in range(num_individuals)]
```

```python
def build_backcross_generations(
    base_name: str,
    initial_hybrid_gen_label: str, # This will be 'F1' (or whatever starts the BC series)
    ancestor_pop_label: str,       # This will be "Ancestry_1" or "Ancestry_2"
    num_backcross_generations: int # How many BC generations do you want (e.g., 5 for BC1, BC2, BC3, BC4, BC5)
) -> List[Tuple[str, str, str]]:
    """
    Builds a sequential backcross generation plan.

    In a backcross, a hybrid generation is repeatedly crossed with one of its
    ancestral (recurrent) populations. This function generates a plan for
    `num_backcross_generations` such crosses, starting from an `initial_hybrid_gen_label`.

    Args:
        base_name (str): The prefix for backcross generation names (e.g., "BC").
        initial_hybrid_gen_label (str): The label of the first hybrid generation that will
                                         be used in the initial backcross (e.g., "F1").
        ancestor_pop_label (str): The label of the ancestral population (e.g., "Ancestry_1" or "Ancestry_2")
                                   that the hybrid generations will be repeatedly crossed with. This is the recurrent parent.
        num_backcross_generations (int): The total number of backcross generations to create
                                         (e.g., 5 for BC1 to BC5).

    Returns:
        List[Tuple[str, str, str]]: A list of backcross generation crosses. Each tuple
                                     represents (new_generation_label, hybrid_parent_label, recurrent_parent_label).
                                     Example: [('BC1_A1', 'F1', 'Ancestry_1'),
                                               ('BC2_A1', 'BC1_A1', 'Ancestry_1'), ...]
    """
    plan = [] # Initialize an empty list to store the sequence of planned backcrosses.
    recurrent_ancestor = ancestor_pop_label # The ancestral population used for all backcrosses.
    current_hybrid_ancestor = initial_hybrid_gen_label # The hybrid generation used in the *current* cross.

    # Define a mapping to create shorter, descriptive labels for the ancestor in the generation name.
    # This aligns with the user's request for formats like 'BC1_A1' or 'BC1_A2'.
    ancestor_short_label_map: Dict[str, str] = {
        'Ancestry_1': 'A1', # Assuming 'Ancestry_1' is associated with allele type '1' (or similar logic)
        'Ancestry_2': 'A2'  # Assuming 'Ancestry_2' is associated with allele type '2' (or similar logic)
    }

    # Get the short label for the current `ancestor_pop_label`.
    # `.get()` is used for safe lookup, providing 'UNKNOWN' if the label isn't mapped.
    ancestor_short_label = ancestor_short_label_map.get(ancestor_pop_label, 'UNKNOWN')

    # Iterate to create the desired number of backcross generations (e.g., BC1, BC2, ..., BCn).
    for i in range(1, num_backcross_generations + 1):
        # Construct the label for the new backcross generation.
        # This combines the base name ('BC'), the generation number (i), an underscore '_',
        # and the short label for the recurrent ancestor (e.g., "BC1_A1").
        backcross_label = f"{base_name}{i}_{ancestor_short_label}"

        # Append the planned cross to the list:
        # (Label of the new generation, Label of the hybrid parent, Label of the recurrent ancestor).
        plan.append((backcross_label, current_hybrid_ancestor, recurrent_ancestor))

        # For the next backcross generation, the newly created backcross generation
        # (e.g., BC1_A1) becomes the hybrid parent that will be crossed again
        # with the recurrent ancestor.
        current_hybrid_ancestor = backcross_label

    return plan # Return the complete list of planned backcrosses.
```

```python
def build_fillial_generations(base_name: str, start_gen: int, end_gen: int) -> List[Tuple[str, str, str]]:
    """
    Creates a crossing plan for sequential filial (F-series) generations.

    This function outlines the crosses needed to produce a series of filial generations,
    starting from a defined generation number (`start_gen`) up to `end_gen` (inclusive).
    The first generation in the sequence is defined as a cross between two distinct
    ancestral populations. All subsequent generations in this series are produced
    by intercrossing individuals from the immediately preceding generation.

    Args:
        base_name (str): The prefix for the generation names (e.g., "F" for Filial generations).
        start_gen (int): The starting generation number for the sequence (e.g., 1 for F1).
        end_gen (int): The final generation number to include in the plan (e.g., 5 for F5).

    Returns:
        List[Tuple[str, str, str]]: A list of tuples, where each tuple represents a planned cross:
                                     (new_generation_label, parent1_label, parent2_label).
                                     Example: [('F1', 'Ancestry_1', 'Ancestry_2'),
                                               ('F2', 'F1', 'F1'),
                                               ('F3', 'F2', 'F2'), ...]
    """
    plan = [] # Initialize an empty list to store the sequence of planned crosses.

    # Iterate through the range of generations from `start_gen` to `end_gen`.
    for i in range(start_gen, end_gen + 1):
        # Construct the label for the current generation (e.g., "F1", "F2", "F3").
        current_gen_label = f"{base_name}{i}"

        if i == start_gen:
            # For the very first generation in the sequence (e.g., F1),
            # the cross is always defined between the two initial ancestral populations.
            plan.append((current_gen_label, 'Ancestry_1', 'Ancestry_2'))
        else:
            # For all subsequent filial generations (e.g., F2, F3, F4...),
            # the cross is defined as individuals from the *previous* generation
            # being intercrossed amongst themselves.
            previous_gen_label = f"{base_name}{i-1}"
            plan.append((current_gen_label, previous_gen_label, previous_gen_label))

    return plan # Return the complete list of planned crosses for the filial generations.
```

```python
class Chromosome:
    """
    Represents a single haploid chromosome strand.

    Each Chromosome object holds a sequence of alleles (integer values)
    corresponding to loci along its length.
    """
    def __init__(self, alleles: List[int]):
        """
        Initializes a Chromosome object with a list of alleles.

        Args:
            alleles (List[int]): A list of integer alleles (0 or 2) representing
                                 the genetic sequence along the chromosome.
        """
        self.alleles = alleles

    def __repr__(self) -> str:
        """
        Provides a concise string representation for the Chromosome object.

        This method is primarily used for debugging and allows you to easily see
        a preview of the chromosome's allele sequence when you print a Chromosome object.
        It shows the first 10 alleles if available.
        """
        snippet = ''.join(map(str, self.alleles[:10])) if self.alleles else ''
        return f"Chr({snippet}...)"
```

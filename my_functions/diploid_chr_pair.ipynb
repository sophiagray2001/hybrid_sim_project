```python
class DiploidChromosomePair:
    """
    Represents a pair of homologous chromosomes.

    This class encapsulates two 'Chromosome' objects, typically representing
    one chromosome inherited from each parent, forming a diploid pair.
    """
    def __init__(self, chromatid1: Chromosome, chromatid2: Chromosome):
        """
        Initializes a DiploidChromosomePair with two Chromosome objects.

        Args:
            chromatid1 (Chromosome): The first chromosome (haploid strand) in the pair.
            chromatid2 (Chromosome): The second chromosome (haploid strand) in the pair.
        """
        self.chromatid1 = chromatid1
        self.chromatid2 = chromatid2

    def __repr__(self) -> str:
        """
        Provides a detailed string representation for the DiploidChromosomePair.

        This method displays the representations of both constituent 'Chromosome' objects,
        making it easy to inspect the pair's composition.
        """
        return f"Pair(\n  {self.chromatid1}\n  {self.chromatid2}\n)"
```

```python
def genotype_numeric(allele_a: int, allele_b: int) -> int:
    """
    Converts two allele integers into a numeric genotype code.

    This function determines the genotype state (homozygous Allele0, homozygous Allele2,
    or heterozygous) based on the numeric values of the two alleles at a given locus.
    It's a pure function, meaning its output depends only on its inputs.

    Args:
        allele_a (int): The numeric value of the first allele (expected to be Allele0 (0) or Allele2 (2)).
        allele_b (int): The numeric value of the second allele (expected to be Allele0 (0) or Allele2 (2)).

    Returns:
        int: An integer representing the genotype:
             - 0 if homozygous Allele0 (both alleles are 0)
             - 2 if homozygous Allele2 (both alleles are 2)
             - 1 if heterozygous (one allele is 0 and the other is 2)
    """
    if allele_a == allele_b:
        return allele_a  # Returns 0 for (0,0) or 2 for (2,2)
    else:
        return 1  # Returns 1 for (0,2) or (2,0)
```

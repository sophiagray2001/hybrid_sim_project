```python
def population_stats(pop: List['Individual']) -> dict:
    """
    Calculates key summary statistics for a given population of 'Individual' objects.

    This helper function provides a quick overview of the genetic composition of a population
    by computing the mean and standard deviation of hybrid index (HI) and heterozygosity (HET),
    as well as the total count of individuals in the population.

    Args:
        pop (List[Individual]): A list of 'Individual' objects representing the population
                                for which statistics are to be calculated.

    Returns:
        dict: A dictionary containing the following summary statistics:
              - 'mean_HI': The average Hybrid Index across all individuals in the population.
              - 'std_HI': The standard deviation of the Hybrid Index.
              - 'mean_HET': The average Heterozygosity across all individuals.
              - 'std_HET': The standard deviation of Heterozygosity.
              - 'count': The total number of individuals in the population.
              Returns 0 for mean/std if the population is empty.
    """
    # Calculate the Hybrid Index for each individual in the population.
    his = [ind.calculate_hybrid_index() for ind in pop]
    # Calculate the Heterozygosity for each individual in the population.
    hets = [ind.calculate_heterozygosity() for ind in pop]

    # Return a dictionary containing the calculated statistics.
    # Conditional checks (e.g., `if his else 0`) prevent errors like division by zero or
    # attempting to calculate statistics on an empty list if the population is empty.
    return {
        'mean_HI': np.mean(his) if his else 0,   # Mean Hybrid Index
        'std_HI': np.std(his) if his else 0,     # Standard Deviation of Hybrid Index
        'mean_HET': np.mean(hets) if hets else 0, # Mean Heterozygosity
        'std_HET': np.std(hets) if hets else 0,   # Standard Deviation of Heterozygosity
        'count': len(pop)                       # Total number of individuals in the population
    }
```

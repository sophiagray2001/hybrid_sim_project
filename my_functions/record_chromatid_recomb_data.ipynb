```python
def record_chromatid_recombination(individual: 'Individual', generation_label: str):
    """
    Records the detailed recombination block data for an individual's chromatids.

    This function utilises the `Individual` object's built-in capability to analyze
    its own chromatids for recombination events (junctions) and the ancestral blocks
    they define. It then enriches this data with the generation label and appends it
    to a global list for later analysis of recombination patterns across the simulation.

    Args:
        individual (Individual): The 'Individual' object whose chromatid recombination data
                                 needs to be extracted and recorded.
        generation_label (str): A string label (e.g., 'F2', 'BC1_A2') to associate with the
                                current generation for data tracking.
    """
    # Retrieve the raw recombination block data for all chromatids belonging to the individual.
    # The `get_chromatid_block_data()` method returns a list of dictionaries, where each
    # dictionary contains analysis results for a single chromatid (e.g., 'total_junctions',
    # 'block_lengths', 'block_alleles', 'individual_id', 'diploid_chr_id', 'chromatid_in_pair').
    chromatid_data = individual.get_chromatid_block_data()

    # Iterate through each dictionary (representing one chromatid's data) in the retrieved list.
    for record in chromatid_data:
        # Add the 'generation' label to each chromatid's record. This is crucial for
        # filtering and analyzing recombination data by generation later on.
        record['generation'] = generation_label
        # Append the now complete record (including generation information)
        # to the global list `all_chromatid_recombination_data`.
        # This list collects all recombination events observed throughout the simulation.
        all_chromatid_recombination_data.append(record)
```

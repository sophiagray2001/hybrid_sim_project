```python
def record_individual_genome(individual: 'Individual', generation_label: str):
    """
    Records the full genotype of each locus for every chromosome pair within a given individual.

    This function iterates through all diploid chromosome pairs and each locus on those chromosomes
    to extract and format the allele information. This detailed genetic data is then stored
    in the global `all_locus_genotype_data` list for later analysis.

    Args:
        individual (Individual): The 'Individual' object whose complete genome data needs to be recorded.
        generation_label (str): A string label (e.g., 'F2', 'BC1_A2') to associate with the current generation
                                for data tracking and filtering.
    """
    # Iterate through each diploid chromosome pair belonging to the individual.
    # `chr_idx` will be 0, 1, 2... corresponding to the chromosome pair index.
    for chr_idx, pair in enumerate(individual.diploid_chromosome_pairs):
        # For each chromosome pair, iterate through all loci (positions) on that chromosome.
        # `locus_idx` will be 0, 1, 2... up to `num_loci_per_chromosome - 1`.
        for locus_idx in range(individual.num_loci_per_chromosome):
            # Extract the allele (numeric value: 0 or 2) from the first chromatid at the current locus.
            allele_a = pair.chromatid1.alleles[locus_idx]
            # Extract the allele (numeric value: 0 or 2) from the second chromatid at the current locus.
            allele_b = pair.chromatid2.alleles[locus_idx]

            # Form a standardized string representation of the genotype for this locus (e.g., "2|0").
            genotype_str = f"{allele_a}|{allele_b}"

            # Append a dictionary containing all the relevant details for this specific locus
            # to the global list `all_locus_genotype_data`.
            # This allows for comprehensive logging of the genetic state of the population.
            all_locus_genotype_data.append({
                'generation': generation_label,     # The generation this individual belongs to
                'individual_id': individual.id,    # The unique ID of the individual
                'diploid_chr_id': chr_idx + 1,     # 1-based index of the chromosome pair for readability
                'locus_position': locus_idx,       # 0-based index of the locus on the chromosome
                'genotype': genotype_str           # The combined allele string for this locus
            })
```
